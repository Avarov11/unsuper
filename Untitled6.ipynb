```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('C:/Users/avrat/Downloads/CC GENERAL.csv')
df
```

```python
df.info()
```

```python
df=df.drop(['CREDIT_LIMIT','MINIMUM_PAYMENTS','CUST_ID','TENURE'],axis=1)
df
```

```python
df
```

```python
x=df[['ONEOFF_PURCHASES_FREQUENCY','PURCHASES_INSTALLMENTS_FREQUENCY']]
x
```

```python
from sklearn.cluster import AgglomerativeClustering #Importing our clustering algorithm : Agglomerative
model=AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='complete')
clust_labels=model.fit_predict(x) 
```

```python
agglomerative=pd.DataFrame(clust_labels)
agglomerative
```

```python
import matplotlib.pyplot as plt
fig =plt.figure()
ax = fig.add_subplot(111)
scatter = ax.scatter (df['ONEOFF_PURCHASES_FREQUENCY'] ,df["PURCHASES_INSTALLMENTS_FREQUENCY"] , c= agglomerative[0], s=50)
plt.colorbar(scatter)
```

```python
from sklearn.cluster import KMeans  #Importing our clustering algorithm: KMeans
kmeans=KMeans(n_clusters=5, random_state=0)  #Cluster our data by choosing 5 as number of clusters
y=kmeans.fit(x)
```

```python
labels=pd.DataFrame(y.labels_)
labels
```

```python
kmeans.predict(x)
print(kmeans.cluster_centers_) 
```

```python

plt.scatter(df["ONEOFF_PURCHASES_FREQUENCY"][y.labels_ == 0],          
            df["PURCHASES_INSTALLMENTS_FREQUENCY"][y.labels_ == 0],s=80,c='magenta',label='Careful')
plt.scatter(df["ONEOFF_PURCHASES_FREQUENCY"][y.labels_ == 1],
           df["PURCHASES_INSTALLMENTS_FREQUENCY"][y.labels_ == 1],s=80,c='yellow',label='Standard')
plt.scatter(df["ONEOFF_PURCHASES_FREQUENCY"][y.labels_ == 2],
           df["PURCHASES_INSTALLMENTS_FREQUENCY"][y.labels_ == 2],s=80,c='green',label='Target')
plt.scatter(df["ONEOFF_PURCHASES_FREQUENCY"][y.labels_ == 3], 
           df["PURCHASES_INSTALLMENTS_FREQUENCY"][y.labels_ == 3],s=80,c='cyan',label='Careless')
plt.scatter(df["ONEOFF_PURCHASES_FREQUENCY"][y.labels_ == 4], 
           df["PURCHASES_INSTALLMENTS_FREQUENCY"][y.labels_ == 4],s=80,c='burlywood',label='Sensible')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', label = 'Centroids')
plt.legend()
plt.show()
```

```python
Sum_of_squared_distances=[]
K = range (1,10)
for k in K:
    km =KMeans (n_clusters=k)
    km =km.fit(df)
    Sum_of_squared_distances.append(km.inertia_)
plt.plot(Sum_of_squared_distances,'ro-')
```

```python
from sklearn.metrics import silhouette_score
from sklearn.metrics import calinski_harabasz_score
from sklearn.metrics import davies_bouldin_score
def metrics(data,labels):
    ss = silhouette_score(data,labels)
    chs = calinski_harabasz_score(data,labels)
    dbs = davies_bouldin_score(data,labels)
    return ss , chs ,dbs
```

```python
km_ss,km_chs,km_dbs=metrics(x,y.labels_)
```

```python
ag_ss,ag_chs,ag_dbs=metrics(x,clust_labels)
```

```python
results={"Agglomerative Clustering":{"silhouette_score":ag_ss,"calinski_harabasz_score":ag_chs,"davies_bouldin_score":ag_dbs},"K-means Clustering":{"silhouette_score":km_ss,"calinski_harabasz_score":km_chs,"davies_bouldin_score":km_dbs}}
results = pd.DataFrame(results)
results.transpose()
```
